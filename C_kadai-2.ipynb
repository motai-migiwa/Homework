# テスト結果データを生成する
- 科目と得点を乱数によって作る

```python
import random
```

### 科目の配列を用意して繰り返しで出力する

```python
#科目のリスト
subject = ["国語", "数学", "地歴", "英語"]
```

```python
for i in subject:
    print(i)
```

```python
subject[1]
```

### ランダム関数を使って0から10までの値を出力する

```python
#　0〜10まででランダムに数を生成する
random.randint(0, 10)
```

### randomを使ってランダムに科目を出力する

```python
#添え字をランダムにして科目を指定
subject[random.randint(0, 3)]
```

### randomを使って点数を生成(0〜100)

```python
#得点をランダムで生成
random.randint(0, 100)
```

```python
data = []
for i in range(1000):
    col = []
    #クラスをランダムに生成して追加
    col.append(random.randint(0, 3))
    #科目をランダムに生成して追加
    col.append(subject[random.randint(0, 3)])
    #得点をランダムに生成して追加
    col.append(random.randint(0, 100))
    #科目・得点がはいっている配列colをdataに追加する
    data.append(col)
```

```python
data
```

## 教科ごとに得点をまとめる

```python
#辞書型の配列を生成する
subject_dict = {"国語":0, "数学":1, "地歴":2, "英語":3}
```

```python
subject_dict['英語']
```

```python
subject_score = [[], [], [], []]
for i in data:
    subject_score[subject_dict[i[1]]].append(i[1])
```

```python
subject_score[0]
```

## 平均・最大値・最小値

```python
#平均を求める
sum(subject_score[0])/len(subject_score[0])
```

```python
#最大値を求める
max(subject_score[0])
```

```python
#最小値を求める
min(subject_score[0])
```

```python
#科目ごとに平均・最大値・最大値
for i in subject_score:
    print("平均: {}".format(sum(i)/len(i)))
    print("最大値: {}".format(max(i)))
    print("最小値: {}".format(min(i)))
```

```python
#pandasのデータフレームに変換
df = pd.DataFrame(data)
```

```python
#科目が数学のデータのみを抽出する
df1 = df[(df[0] == 2) & (df[1] == '英語') ]
```

```python
#平均・最大・最小を出力する関数
def df_output(class_num, sub):
    df1 = df[(df[0] == class_num) & (df[1] == sub)]
    print("平均: {}".format(sum(df1[2])/len(df1[2])))
    print("最大値: {}".format(max(df1[2])))
    print("最小値: {}".format(min(df1[2])))
```

```python
def output(subj):
    for i in range(0, 3):
        print("{}組の{}の成績".format(i, subj))
        df_output(i, subj)
        print()
```

```python
output("数学")
```
