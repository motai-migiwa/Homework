```python
date = []
for i in range(1000):
    col = []
    col.append(random.randint(0,3))
    col.append(subject[random.randint(0,3)])
    col.append(random.randint(0,100))
    date.append(col)
```

```python
date
```

```python
subject_dict = {'国語':0,'数学':1,'地歴':2,'英語':3}
```

```python
subject_dict['英語']
```

```python
subject_score = [[],[],[],[]]
for i in date:
    subject_score[subject_dict[i[1]]].append(i[1])
```
